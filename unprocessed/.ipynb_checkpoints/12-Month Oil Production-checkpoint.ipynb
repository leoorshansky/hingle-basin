```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os.path import join, dirname, isfile, basename
from os import getcwd, listdir
import math as math
import os
```

```python
unprocessed_dir = join(join(dirname(getcwd()), 'unprocessed'), 'well production')
unprocessed_dir
```

```python
data = pd.read_csv(join(unprocessed_dir, 'well production.csv'))
```

```python
oil_sums = []
for row in data.index:
    well_oil_sum = 0
    for i in range(1, 13):
        well_oil_sum = well_oil_sum + data.iloc[row]['oil ' + str(i)]
    oil_sums.append([data.iloc[row]['well name'], well_oil_sum])
oil_sums = pd.DataFrame(oil_sums, columns = ['Well Name', 'Annual Oil Output'])
oil_sums = oil_sums.sort_values(by='Well Name')
oil_sums
```

```python
northings = []
eastings = []
n_count = 0
e_count = 0
for filename in os.listdir(unprocessed_dir):
    if filename.endswith(".csv") and not "well production" in filename:
        well_data = pd.read_csv(join(unprocessed_dir, filename))
        northings.append(well_data.iloc[0]['northing'])
        eastings.append(well_data.iloc[0]['easting'])
oil_sums['Northing'] = northings
oil_sums['Easting'] = eastings
oil_sums
```

```python
all_data = pd.read_csv(join(join(dirname(getcwd()), 'mapping'), 'all_data.csv'))
```

```python
all_data['12-month oil production'] = 0
```

```python
all_data
```

```python
for i in all_data.index: 
    for j in oil_sums.index:
        if (all_data.iloc[i]['northing'] == oil_sums.iloc[j]['Northing'] and all_data.iloc[i]['easting'] == oil_sums.iloc[j]['Easting']):
            print('found matching position')
            all_data[i]['12-month oil production'] = oil_sums.iloc[j]['Annual Oil Output']
```
