```python
import numpy as np
import pandas as pd
from os.path import join, dirname, isfile, basename
from os import getcwd, listdir
import matplotlib.pyplot as plt
```

```python
import sys
if 'jupyterthemes' in sys.modules:
    from jupyterthemes import jtplot
    jtplot.style()
```

```python
base_path = dirname(getcwd())
data_path = join(join(base_path, 'mapping'), 'all_data.csv')
all_data = pd.read_csv(data_path).filter(['easting', 'northing', 'oil saturation', 'porosity'])
```

```python
all_data['OOIP'] = np.log10(all_data['oil saturation'] * all_data['porosity'] * 7758)
```

```python
plt.figure()
c = all_data['OOIP']
plt.scatter(all_data['easting'], all_data['northing'], c=c)
plt.show()
```
